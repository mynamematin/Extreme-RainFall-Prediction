```python
import xarray as xr
import numpy as np
import pandas as pd
import scipy as sc
from scipy.signal import butter, filtfilt
import matplotlib.pyplot as plt
import seaborn as sns

add_data = "/home/cccr/roxy/matin/MTech_project/data/"
```

```python
dataset = xr.open_dataset( add_data + "sst40yearmean.nc")
ds = xr.open_dataset( add_data + "sst.day.mean.2021.nc")
```

```python


# Calculate the SST mean
sst_mean = dataset.sst.mean(dim=['time'])

# Set up the figure and axes
fig, ax = plt.subplots(figsize=(8, 6))

# Create a contour plot of the SST mean
contour = ax.contourf(sst_mean.lon, sst_mean.lat, sst_mean, cmap='coolwarm')

# Add a colorbar to the plot
cbar = fig.colorbar(contour)

# Set the title and axis labels
ax.set_title('SST 40 year Mean 1982-2021')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Show the plot
plt.savefig("SST 40 year Mean 1982-2021",dpi=300)
plt.show()
```
